Following columns are not useful to us and thus we can drop them
* <b>EmployeeCount</b>
* <b>StandardHours</b>
* <b>EmployeeNumber</b>

```python
num_col = list(df.describe().columns)
col_categorical = list(set(df.columns).difference(num_col))
remove_list = ['EmployeeCount','StandardHours','EmployeeNumber']
col_numerical = [e for e in num_col if e not in remove_list]
```

```python
print(len(num_col))
print(len(col_categorical))
print(len(col_numerical))
```

```python
df[col_numerical].corr()
```

```python
plt.figure(figsize=(24,8))
sns.heatmap(df[col_numerical].corr(), annot=True, fmt=".2f")
```

```python
plt.figure(figsize=(24,8))
mask = np.zeros_like(df[col_numerical].corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.light_palette((210,90,60),input= "husl")
sns.heatmap(df[col_numerical].corr(), mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink"})
```
